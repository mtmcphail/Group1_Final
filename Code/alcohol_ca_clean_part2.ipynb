### Check SQL database (full dataframe) and delete unncessary columns.

```python
df= df.drop(columns = 'Unnamed: 0', axis=1)
df
```

### Sum the binned race/ethnic categories from full dataframe

```python
# add geoname with new csv file
df_sum = df.groupby(['geotype', 'geoname', 'county_name','race_eth_updt'], 
                    as_index=False).agg({'num_people_qrt': 'sum', 
                                         'tot_people': 'sum',
                                         'region_name' : 'first',
                                         'per_capita_income' : 'first',
                                         'total_income' : 'first',
                                         'population' : 'first',
                                         'unemp_2014' : 'first',
                                         'median_house_price' : 'first',
                                         'median_house_pct' : 'first',
                                         'metro_status' : 'first',
                                         'economic_type_label' : 'first',
                                         'low_education_2015' : 'first',
                                         'low_employment_cnty' : 'first',
                                         'pop_loss_2010' : 'first',
                                         'retirement_dest_2015' : 'first',
                                         'persistent_poverty_2013' : 'first',
                                         'persistent_child_poverty_2013' : 'first'})
 
df_sum    
```

### Merge new full dataframe (df_sum) with totals datframe (df_total) to calculate new percentages

```python
##  need to update with geoname
df_new = pd.merge(df_sum, df_total, on=['geotype', 'geoname', 'county_name'], how="left")
df_new.info()
```

```python
# re-order columns
df_new = df_new[['county_name',
                'geotype', 
                'geoname',
                'race_eth_updt',
                'num_people_qrt', 
                'tot_people', 
                'tot_people_qrt',
                'tot_people_region',
                'region_name',
                'per_capita_income',
                'total_income', 
                'population',
                'unemp_2014',
                'median_house_price',
                'median_house_pct',
                'metro_status',
                'economic_type_label',
                'low_education_2015', 
                'low_employment_cnty',  
                'pop_loss_2010',  
                'retirement_dest_2015', 
                'persistent_poverty_2013',
                'persistent_child_poverty_2013']]

#rename columns
df_new = df_new.rename(columns = {'num_people_qrt' : 'num_ppl_qrt_race',
                                  'tot_people' : 'tot_ppl_region_race'}, inplace = False)
df_new
```

### Calculate % of total race population and % of total pupolation for each race

```python
# of the race/ethnic population living within a region, what percentage of the race/ethnicity lives within 0.25
df_new['pct_ppl_qrt_race'] = df_new['num_ppl_qrt_race'] / df_new['tot_ppl_region_race']

# of the total population living within a region within 0.25 miles, 
# what percentage is a given
df_new['pct_all_qrt_race'] = df_new['num_ppl_qrt_race'] / df_new['tot_people_qrt']

# of the total population living within a region, what percentage of a given race lives within 0.25 miles 
df_new['pct_all_ppl_qrt_race'] = df_new['tot_people_qrt'] / df_new['tot_people_region']


#df_state_pct['pct_total'] = df_state_pct['people_qtr_sum'] / df_state_pct['tot_people']
df_new.info()
```

```python
# make a copy
df_final = df_new

# convert Nan percentages (where denominator = 0) to 0.
df_final['pct_ppl_qrt_race'] = df_final['pct_ppl_qrt_race'].replace(np.nan, 0)
df_final['pct_all_qrt_race'] = df_final['pct_all_qrt_race'].replace(np.nan, 0)
df_final['pct_all_ppl_qrt_race'] = df_final['pct_all_ppl_qrt_race'].replace(np.nan, 0)

df_final.info()
```
